# Importing Input data 

```python
input=pd.read_csv(r"C:\Users\vasan\Downloads\Input.xlsx - Sheet1.csv")
input
```

# Creating a loop to automatically fetch and save the data in csv file 

```python

for i in input.values:
    url=i[1]
    source=requests.get(url)
    if (source.status_code==200):
        print('DataFetched',i[0])
        soup=BeautifulSoup(source.text,'lxml')
        article=soup.find('article')
        for article in soup.find_all('article'):
            headline=article.h1.text
            summary=article.find('div',class_="td-post-content").text
            summary=summary.replace('\n',' ')
            csv_file=open(f'{i[0]}.csv','w',encoding='utf-8')
            csv_writer=csv.writer(csv_file)
            csv_writer.writerow(['headline','summary'])
            csv_writer.writerow([headline,summary])
            csv_file.close()
    else:
        print('Not Working',i[0])
        
```
